## ***Data Wrangling***

### Outlier removal And feature engineering

```python
# Define function to check for outliers
def out_dec(i):
  sns.boxplot(x=df[i])
  plt.title(f'Boxplot showing outliers')
  plt.xlabel(f'{i}')
  plt.show()
```

```python
# Checking for outliers on price column
out_dec('price')
```

Due to extreme outliers we will remove to five percentile of the price column.

```python
# Set top 5 percentile
percentile_5 = df['price'].quantile(0.95)

# Clip values at the 5th percentile
df['price'] = df['price'].clip(upper=percentile_5)
```

```python
# Checking for outliers again
out_dec('price')
```

```python
df.columns
```

```python
#We will create new column for price per night
df['price_per_night']=df['price']/df['minimum_nights']
```

## **Understand the relationships between variables**

#### Univariate analysis

For univariate analysis we will use KDE plots for continous variables, and countplots for categorical variables.

```python
# Chart - 1 visualization code
sns.kdeplot(df[df['price']>0]['price'] , fill=True)
plt.vlines( np.median(df[df['price'] > 0]['price']), ymin=0, ymax=0.009, color='red', linestyle='dashed', linewidth=2)
plt.vlines( np.mean(df[df['price'] > 0]['price']), ymin=0, ymax=0.009, color='blue')
plt.title('Price KDE')
plt.show()
```

Despite removing some outliers, since mean is higher than median, this dataset is still right skewed.

```python
#We will look at neighbourhood_group
sns.countplot(x=df['neighbourhood_group'])
plt.xlabel('Neighbourhood Group')
plt.title('Countplot for neighhbourhood')
plt.show()
```

As we can see, Brooklyn and Manhattan form vast majority of this dataset.

```python
#Countplot for Room Type
sns.countplot(x=df['room_type'])
plt.xlabel('Room Type')
plt.title('Countplot for Room Type')
plt.show()
```

As we can see shared room type arrangements seem to be in the minority, while private rooms and entire home apartments are almost equal in number.

```python
#Graph to show top ten hosts
plt.figure(figsize=(12,5))
top_ten_hosts=df['host_name'].value_counts()[:10]
top_ten_hosts.plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```

Looks like the top hosts rent out hundreds of properties.

```python
#Chart for
plt.figure(figsize=(20,5))
top_neigh=df['neighbourhood'].value_counts(ascending=False)[:10]
top_neigh.plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```

The top neighbourhoods have thousands of properties

#### Bivariate Analysis

For bivariate analysis we will use a combination of countplot and violinplots to compare variables.

```python
# Checking relationship between price and room type
sns.violinplot(x=df['price'],y=df['room_type'],fill=True)
plt.xlabel('Price')
plt.ylabel('Room Type')
plt.title('Price distribution for each room type')
plt.show()
```

As expected entire home/apartment are significantly pricier.

```python
# Checking relationship between price and Neighbourhood Group
plt.figure(figsize=(10,7))
sns.violinplot(x=df['price'],y=df['neighbourhood_group'], fill=True)
plt.xlabel('Price')
plt.ylabel('Neighbourhood Group')
plt.title('Price distribution for each Neighbourhood Group')
plt.show()
```

Manhattan seem to have the most expensive properties to rent.

```python
# Room type breakdown for each neighbourhood group
sns.countplot(x=df['neighbourhood_group'],hue=df['room_type'])
plt.xlabel('Neighbourhood Group')
plt.title('Room type for every neighbourhood group')
plt.show()
```

Unlike every other neighbourhood group, in Manhattan entire home apartments outnumber every other type. This most likely explains the price differential.

#### Multivariate analysis

For multivaiate analysis we will use barplots with subcategories and scatterplots for geolocational analysis.

```python
# Comparing per night cost of each neighbouhood group
plt.figure(figsize=(10,6))
sns.barplot(x=df['neighbourhood_group'],y=df['price_per_night'],hue=df['room_type'],estimator=np.median, errorbar=None)
plt.show()
```

Looks like median prices are higher in Manhattan irrespective of room types.

```python
# Map showing properties to rent in the entire NYC
plt.figure(figsize=(8,8))
sns.scatterplot(x=df['latitude'],y=df['longitude'],hue=df['neighbourhood_group'],size=df['price'], alpha=0.4)
plt.show()
```

At on glance, we can see the differences in property density and price differential.

```python
# Dropping irrelevant columns for correlation matrix
subset_df = df.drop(['id', 'host_id', 'latitude', 'longitude'], axis=1)
```

```python
# Compute the correlation matrix
corr = subset_df.select_dtypes('number').corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(8, 8))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.xticks(rotation=90)
plt.show()
```

Outside of thsome obvious examples, none of the variables are strongly correlated.

## **How to fullfill Business Objective**

#### How should Airbnb maximize profit ?

1. Pricing Strategy: Since the dataset is right-skewed, consider using the median price instead of the mean when setting base prices. This could help avoid overpricing caused by outliers. Also, as you’ve observed, entire homes/apartments are significantly pricier. You could consider offering discounts or promotions for these types of listings to increase their affordability and attract more customers.

2. Targeted Marketing: Brooklyn and Manhattan form the majority of this dataset, so consider focusing your marketing efforts in these areas. You could also use the neighbourhood data to identify areas with high demand and low supply, and encourage more listings in these areas.

3. Service Offerings: Shared room type arrangements seem to be in the minority, while private rooms and entire home apartments are almost equal in number. This suggests that customers prefer private spaces. Consider promoting private rooms and entire homes/apartments more heavily in your marketing materials.

4. Geolocation Analysis: Manhattan seems to have the most expensive properties to rent, and entire home apartments outnumber every other type in this area. This could be a sign of high demand for entire homes/apartments in Manhattan. Consider using this information to adjust your pricing strategy in this area. Also, since median prices are higher in Manhattan irrespective of room types, consider offering more competitive pricing in this area to attract more customers.
